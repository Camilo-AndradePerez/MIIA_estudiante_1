```python

# código

# importación de paquetes necesarios
import math
import matplotlib as mp
import numpy as np


# función que calcula la media de una lista de datos numéricos

def media(datos):
    suma = 0
    for i in datos: 
        suma = suma + float(i)
    media = suma/len(datos)    
    return media

media(promedio)


# función que calcula la mediana de una lista de datos numéricos

def mediana(datos):
    lista_ordenada = sorted(datos)
    if len(datos)%2 == 0:
        valor_inicial = float(lista_ordenada[int(len(datos)/2)])
        #print("Longitud Datos: ", int(len(datos)/2))
        #print("Valor Inicial: ", valor_inicial)
        valor_final = float(lista_ordenada[int(len(datos)/2+1)])
        #print("Valor Final: ", valor_final)
        mediana = (valor_inicial + valor_final)/2
    else:
        posicion = int((len(datos)/2)+1)
        mediana = lista_ordenada[posicion]
        
    return mediana

mediana(estrato)


# función que calcula la desviación de una lista de datos numéricos

def desviacion_estandar(datos):
    suma_m = 0
    for i in datos: 
        suma_m = suma_m + float(i)
    media = suma_m/len(datos)    
    
    suma_d = 0
    for i in datos: 
        suma_d = suma_d + pow((float(i)-float(media)),2)
    desv_est = pow((suma_d/len(datos)),0.5)
    return desv_est

desviacion_estandar(promedio)


# función que calcula la varianza de una lista de datos numéricos

def varianza(datos):
    suma_m = 0
    for i in datos: 
        suma_m = suma_m + float(i)
    media = suma_m/len(datos)    
    
    suma_d = 0
    for i in datos: 
        suma_d = suma_d + pow((float(i)-float(media)),2)
    varianza = (suma_d/len(datos))
    return varianza

varianza(edad)


#función que crea conteos a partir de listas de texto

def frecuencias_lista(lista):
    frecuencia = {}
    for n in lista:
        if n in frecuencia:
            frecuencia[n] += 1
        else:
            frecuencia[n] = 1
    print("Los valores únicos de la lista y su respectiva frecuencia son: ", frecuencia) 
    print()
    #import matplotlib.pyplot as plt
    #fig, ax = plt.subplots()    
    #freq = ax.bar(lista)  
    #ax.grid(True)  
    #plt.show()        
    return frecuencia
            
frecuencias_lista(escolaridad)


#función que visualiza las características de una población dada
#nota que la variable 'poblacion' es local; podemos pasar cualquier población a la función)

def caracteristicas(datos):
    print("La cantidad de registros de la lista de datos es: ", len(datos))
    print("El valor mínimo de los datos es: ", min(datos))
    print("El promedio de los datos es: ", media(datos))
    print("La mediana de los datos es: ", mediana(datos))
    print("El valor máximo de los datos es: ", max(datos))
    print("La desviación estándar de los datos es: ", desviacion_estandar(datos))
    print("La varianza de los datos es: ", varianza(datos))
    
    import matplotlib.pyplot as plt
    fig, ax = plt.subplots()    
    hist = ax.hist(datos)  
    ax.grid(True)  
    plt.show()           
    
    return



```
